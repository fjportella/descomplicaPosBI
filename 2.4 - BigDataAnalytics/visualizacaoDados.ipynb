```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.xlabel("Tempo(s))")
plt.ylabel("Amplitude (m)")
plt.title("Gráfico de Seno")

plt.show()
```

```python
import matplotlib.pyplot as plt

#Gerar alguns dados para plotar

x = [1,2,3,4,5]
y = [6,7,8,9,10]

#Criar um gráfico de linhas usando os dados gerados
plt.plot(x,y)
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("Gráfico Linear Simples")

#Exiba do gráfico usando o comando plt.show
plt.show()
```

```python
#Gráfico de Barras

import matplotlib.pyplot as plt
import numpy as np

categorias = ['Categoria 1', "Categoria 2", "Categoria 3"]
valores = [10,20,30]

plt.bar(categorias, valores)
plt.xlabel("Categorias")
plt.ylabel("Valores")
plt.title("Exemplo de Gráfigo de Barras")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.scatter(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Exemplo de Gráfigo de Dispersão")
plt.show()
```

Aleatoriedade

```python
import random

rand_int = random.randint(1,100)
print("Número aleatório gerado: ", rand_int)
```

```python
import random

rand_list = [random.randint(1,100) for i in range(10)]
print("Lista de números aleatórios geradas: ", rand_list)
```

```python
import random

nomes = ["George", "Maria", "Rafaela", "Samuel", "Atena"]
rand_nomes = random.sample(nomes, 3)
print("Nomes aleatórios gerados", rand_nomes)
```
